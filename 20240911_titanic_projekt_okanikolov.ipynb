2.) **Analyse und Visualisierung**


2.1.1 **Personenklasse (Pclass)** - Tabellarische Ansicht

```python

pclass_analyse = titanic.groupby(['pclass', 'survived'])['fare'].mean().unstack()
display(pclass_analyse)
oka
```

2.1.2. **Personenklasse (Pclass)**- Grafische Ansicht

Hier sind die **zusammengefassten Ergebnisse der Analyse** des **Titanic-Datensatzes** in Bezug auf **„Survived“ (Überlebensstatus)** und **„Fare“ (Ticketpreis)** unter Berücksichtigung der Variablen „Pclass“, „Sex“, „Age“, „Embarked“ und „Family Size“:

**1. Personenklasse (Pclass)**
- Tabellarische Ergebnisse: Der durchschnittliche Ticketpreis („Fare“) variiert je nach Personenklasse („Pclass“).
Höhere Klassen (1. Klasse) haben im Durchschnitt höhere Ticketpreise als niedrigere Klassen (2. und 3. Klasse).
Überlebende haben tendenziell höhere durchschnittliche Ticketpreise als Verstorbene, insbesondere in der 1. und 2. Klasse.

- Grafische Ergebnisse: Boxplot zeigt, dass der Ticketpreis in höheren Klassen (1. Klasse) bei Überlebenden höher ist als bei Verstorbenen, während der Unterschied in der 3. Klasse geringer ist.

**2. Geschlecht (Sex)**
- Tabellarische Ergebnisse: Der durchschnittliche Ticketpreis variiert nach Geschlecht („Sex“). Frauen zahlen im Durchschnitt mehr als Männer.
Überlebende Frauen zahlen tendenziell höhere Ticketpreise als männliche Überlebende.
Männer, die überlebt haben, zahlen im Durchschnitt weniger als Männer, die nicht überlebt haben.

- Grafische Ergebnisse: Boxplot zeigt, dass Frauen höhere Ticketpreise zahlen als Männer und dass Überlebende tendenziell höhere Preise bezahlen als Verstorbenen, unabhängig vom Geschlecht.

**3. Alter (Age)**
- Tabellarische Ergebnisse: Der durchschnittliche Ticketpreis variiert je nach Altersgruppe. Jüngere Passagiere (z.B. 0-12 Jahre) zahlen im Durchschnitt weniger.
Die Überlebensrate ist bei jüngeren Passagieren tendenziell höher, und deren Ticketpreise sind im Durchschnitt niedriger.

- Grafische Ergebnisse: Boxplot zeigt, dass jüngere Passagiere tendenziell niedrigere Ticketpreise zahlen, und Überlebende sind häufiger in den jüngeren Altersgruppen vertreten.

**4. Abreisehafen (Embarked)**
- Tabellarische Ergebnisse: Der durchschnittliche Ticketpreis variiert je nach Abreisehafen („Embarked“). Passagiere, die in Southampton („S“) abgefahren sind, zahlen im Durchschnitt höhere Ticketpreise.
Überlebende und Verstorbene haben unterschiedliche Ticketpreise je nach Abreisehafen.

- Grafische Ergebnisse: Boxplot zeigt Unterschiede im Ticketpreis nach Abreisehafen. Passagiere, die aus „S“ (Southampton) abgefahren sind, zahlen tendenziell höhere Ticketpreise.

**5. Familiengröße (Family Size)**


- Tabellarische Ergebnisse: Der durchschnittliche Ticketpreis variiert je nach Familiengröße („Family Size“). Passagiere mit größeren Familien zahlen oft weniger für ihr Ticket.
Überlebende haben tendenziell unterschiedliche Ticketpreise je nach Familiengröße, wobei kleinere Familien häufig höhere Ticketpreise zahlen.

- Grafische Ergebnisse: Boxplot zeigt, dass größere Familien tendenziell niedrigere Ticketpreise zahlen und Überlebende in der Regel unterschiedliche Ticketpreise je nach Familiengröße haben.
